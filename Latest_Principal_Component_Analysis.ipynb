```python
import pandas as pd
import numpy as np
from tabulate import tabulate
from numpy.linalg import eig
train = pd.read_csv('https://raw.githubusercontent.com/gibsonjackson/FODS-Assignment-2/main/Train.csv')
m=len(train)
```

```python
y_train=train['Appliances'].tolist()
```

```python
train = (train - train.mean())/train.std()
```

```python
train=train.drop(columns = 'Appliances')
```

```python
train.head()
```

```python
train_mat=train.to_numpy()
```

```python
train_mat
```

```python
every_feature = np.transpose(train_mat)
```

```python
print(every_feature)
```

```python
covariance_matrix=np.cov(train_mat,rowvar=False)
```

```python
eigen_values = np.linalg.eigvals(covariance_matrix)
```

```python
eigen_values
```

```python
sorted_index = np.argsort(eigen_values)[::-1]
 
sorted_eigenvalue = eigen_values[sorted_index]
```

```python
sorted_index
```

```python
#Calculating Variance
all_variance=[]
sum_variance=np.sum(eigen_values)
for i in range(0,26):
  sum = 0.0
  for j in range(0,i+1):
    sum += eigen_values[j]
  all_variance.append(sum/sum_variance)

print(all_variance)
```

```python
def hypothesis(coeff,rowcount,feature_index):
  y_pred = 0.0
  for each_coeff,each_index in zip(coeff,feature_index):
    y_pred += each_coeff*every_feature[each_index][rowcount]
  return y_pred
```

```python
def PCR(feature_index):
  coeff = np.zeros(len(feature_index))
  for i in range(1,10000):
    rowcount = 0
    for y_actual in y_train:
      y_pred = hypothesis(coeff,rowcount,feature_index)
      feature_matrix=np.zeros(len(feature_index))
      ind = 0
      for each_index in feature_index:
        feature_matrix[ind] = every_feature[each_index][rowcount]
        ind += 1
      coeff=coeff-0.001*(y_pred-y_actual)*feature_matrix
      rowcount += 1
  return coeff
```

```python
all_coeff=[]
for i in range(0,26):
  coeff = PCR(sorted_index[0:i+1])
  all_coeff.append(coeff)
  print(all_coeff)
  print(i)

print(tabulate(all_coeff))
```

```python
def cost(no_of_features,feature_index,coeff,curr_y):
  loss = 0.0
  rowcount = 0
  for y_pred in curr_y:
    feature_matrix=[0.0]*no_of_features
    index = 0
    for each in feature_index:
      feature_matrix[index]=every_feature[each][rowcount]
      index += 1
    rowcount += 1
    loss += (0.5/m)*(np.multiply(coeff,np.transpose(feature_matrix))-y_pred)**2
  sum_loss=0
  for each_loss in loss:
    sum_loss += each_loss
  avg_loss = sum_loss/m
  return avg_loss
  return loss
```

```python
all_train_loss=[]
for i in range(0,26):
  loss = cost(i+1,sorted_index[0:i+1],all_coeff[i],y_train)
  all_train_loss.append(loss)

print(all_train_loss)
```
